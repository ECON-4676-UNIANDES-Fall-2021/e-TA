## **Dictionary based sentiment analysis**

```python
nltk.download('opinion_lexicon')
```

```python
from nltk.corpus import opinion_lexicon
```

```python
pos_list=set(opinion_lexicon.positive())
neg_list=set(opinion_lexicon.negative())
```

```python
def sentiment_dictionary(x):
    pos=0
    neg=0
    for i in x:
        if i in pos_list:
            pos+=1
        if i in neg_list:
            neg+=1
    la={}
    la['positive']=(pos/len(x))*100
    try:
        la['negative']=(neg/len(x))*100
    except:la['negative']=0
    return la
```

```python
sentiment_dictionary(rev1[0])
```

```python
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores(rev[0])
```

```python
p=[]
for i in range(0,len(rev)):
    a=sia.polarity_scores(rev[i])
    b=sentiment_dictionary(rev1[i])
    compa={}
    if a['pos']>=a['neg']:
        compa['sia']=1
    else:
        compa['sia']=0
    if b['positive']>=b['negative']:
        compa['dic']=1
    else:
        compa['dic']=0
    p.append(compa)
    
    
```

```python
data=pd.DataFrame(p)
```

```python
data['res']=data['sia']-data['dic']
```

```python
data.groupby('res').agg({'res':'count'})
```
